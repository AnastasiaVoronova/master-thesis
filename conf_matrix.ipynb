```python
df['category'] = df.category.apply(lambda x: categories_from_indices[categories[x]])
```

```python
score_dummies = pd.get_dummies(df['score'])
category_dummies = pd.get_dummies(df['category'])
df_encoded = pd.concat([score_dummies, category_dummies], axis=1)
corr_matrix = df_encoded.corr().loc[score_dummies.columns, category_dummies.columns]

plt.figure(figsize=(20, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.xlabel("category")
plt.ylabel("score")
plt.show()
```

```python
score_dummies = pd.get_dummies(df['score'])

df_encoded = pd.concat([df['category'], score_dummies], axis=1)
category_share = df_encoded.groupby('category').mean()

plt.figure(figsize=(20, 6))
sns.heatmap(category_share.T, annot=True, cmap='coolwarm', fmt='.2f')
plt.xlabel("category")
plt.ylabel("score")
plt.show()
```
